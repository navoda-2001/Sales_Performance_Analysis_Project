```python
import pandas as pd

data = {
    "Month": ["Jan", "Jan", "Feb", "Feb", "Mar", "Mar"],
    "Region": ["Colombo", "Kandy", "Colombo", "Kandy", "Colombo", "Kandy"],
    "Product": ["Laptop", "Phone", "Laptop", "Phone", "Laptop", "Phone"],
    "Sales": [120000, 80000, 150000, 90000, 170000, 95000]
}

df = pd.DataFrame(data)
df
```

```python
df["Sales"].sum()
```

```python
df["Sales"].mean()
```

```python
df["Sales"].max()
```

```python
df["Sales"].min()
```

```python
df.groupby("Region")["Sales"].sum()
```

```python
df.groupby("Month")["Sales"].sum()
```

```python
df.groupby("Product")["Sales"].sum()
```

```python
import matplotlib.pyplot as plt

df.groupby("Month")["Sales"].sum().plot(kind="bar")
plt.title("Monthly Sales")
plt.ylabel("Revenue")
plt.show()
```

```python
df.groupby("Region")["Sales"].sum().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales Distribution by Region")
plt.ylabel("")
plt.show()
```

INSIGHTS

1. Colombo generated the highest total sales.
2. Sales show a steady increase from January to March.
3. Laptop sales outperform phone sales.
4. March is the best-performing month.
